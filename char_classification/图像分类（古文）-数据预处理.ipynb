```python
import torch
from torch.utils.data import Dataset
from PIL import Image

class GuWenDataset(Dataset):
    def __init__(self, img_paths, transform=None):
        self.img_paths = img_paths
        self.transform = transform

    def __len__(self):
        return len(self.img_paths)

    def __getitem__(self, idx):
        img_path = self.img_paths[idx]
        image = Image.open(img_path).convert('RGB')
        if self.transform:
            image = self.transform(image)

        if "jiagu" in img_path:
            label = 0
        elif "jinwen" in img_path:
            label = 1
        else: # "xiaozhuan" in img_path
            label = 2
            
        return image, label
```

```python
import torch.nn as nn

class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()
        self.conv1 = nn.Conv2d(3, 32, 3)  # 假设输入图片为RGB图, size=(32, 32)
        self.conv2 = nn.Conv2d(32, 64, 3)
        self.fc1 = nn.Linear(64*6*6, 128)  # 参数需根据最终conv2输出的size进行调整
        self.fc2 = nn.Linear(128, 10)  # 假设有10个类别

    def forward(self, x):
        out = F.relu(self.conv1(x))
        out = F.max_pool2d(out, 2)
        out = F.relu(self.conv2(out))
        out = F.max_pool2d(out, 2)
        out = out.view(out.size(0), -1)  # Flatten
        out = F.relu(self.fc1(out))
        out = self.fc2(out)
        return out
```

```python
from torch.utils.data import DataLoader
import torchvision.transforms as transforms
import torch.optim as optim
import torch.nn.functional as F

# 初始化数据集和数据加载器
all_images = jiagu_list + xiaozhuan_list + jinwen_list
transform = transforms.Compose([transforms.Resize((32,32)), transforms.ToTensor()])
dataset = GuWenDataset(all_images, transform=transform)
dataloader = DataLoader(dataset, batch_size=32, shuffle=True)

# 初始化模型、优化器、损失函数
model = CNN()
optimizer = optim.Adam(model.parameters(), lr=0.001)
loss_fn = nn.CrossEntropyLoss()
```
